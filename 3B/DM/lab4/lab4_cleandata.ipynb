```python
import pandas as pd
path="/content/sample_data/__Mall_Customers_toCleanAssignment (1).csv"
dataset = pd.read_csv(path,encoding_errors='ignore')
dataset.rename(columns={"Spending Score (1-100)":"Spending_Score_1_100", "Annual Income (k$)":"Annual_Income_k_Dollar"},inplace=True)
```

```python
df1 = dataset.drop('CustomerID',axis=1)
df1.head()
```

```python
df2=df1.copy()
df2.Gender=df2.Gender.astype(str)
df2.Gender.unique()

df2.loc[df2.Gender=='1','Gender']='Male'
df2.loc[df2.Gender=='0','Gender']='Female'
df2.Gender.unique()

df2.where((df2.Gender=='Male')| (df2.Gender=='Female'),other=None,inplace=True)
df2.Gender.unique()

df2.drop(index=df2[df2.Gender.isnull()].index,inplace=True)
df2.Gender.unique()
```

```python
df3=df2.copy()
df3.describe()

# df3.Spending_Score_1_100.fillna(df3.Spending_Score_1_100.mean(),inplace=True)
df3.Spending_Score_1_100=df3.Spending_Score_1_100.fillna(df3.Spending_Score_1_100.mean())
df3.isnull().sum()

df3.Spending_Score_1_100[df3.Spending_Score_1_100 <1]
df3.Spending_Score_1_100[df3.Spending_Score_1_100 >100]
#
# df3.Spending_Score_1_100[df3.Spending_Score_1_100 <1]=abs(df3.Spending_Score_1_100[df3.Spending_Score_1_100 <1])
# df3.Spending_Score_1_100[df3.Spending_Score_1_100 >100]=df3.Spending_Score_1_100[df3.Spending_Score_1_100 >100]-100
df3.loc[df3.Spending_Score_1_100 <1,'Spending_Score_1_100']=abs(df3.Spending_Score_1_100[df3.Spending_Score_1_100 <1])
df3.loc[df3.Spending_Score_1_100 >100,'Spending_Score_1_100']=df3.Spending_Score_1_100[df3.Spending_Score_1_100 >100]-100
df3.describe()

```

```python
df4=df3.copy()
df4.describe()
# df4.Annual_Income_k_Dollar.fillna(df4.Annual_Income_k_Dollar.mean(),inplace=True)
df4.Annual_Income_k_Dollar=df4.Annual_Income_k_Dollar.fillna(df4.Annual_Income_k_Dollar.mean())
df4.isnull().sum()

df4.Annual_Income_k_Dollar[df4.Annual_Income_k_Dollar <1]
# df4.Annual_Income_k_Dollar[df4.Annual_Income_k_Dollar <1]=abs(df4.Annual_Income_k_Dollar[df4.Annual_Income_k_Dollar <1])
df4.loc[df4.Annual_Income_k_Dollar <1,'Annual_Income_k_Dollar']=abs(df4.Annual_Income_k_Dollar[df4.Annual_Income_k_Dollar <1])
df4.describe()
```

```python
df5=df4.copy()
df5.isnull().sum()
df5 = df5.dropna()
df5.isnull().sum()

df5.Age[df5.Age <1]
df5.loc[df5.Age <1,'Age']=abs(df5.Age[df5.Age <1])

df5.describe()
df5[df5.Age < 15].count()

# Drop if age <15 because they are childern not working
df5.drop(index=df5[df5.Age < 15].index,inplace=True)
df5[df5.Age < 15].count()

df5.describe()

df5.Age=df5.Age.astype(int)
print(df5.Age.dtype)
df5.info()
```
